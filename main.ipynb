```python
import pandas as pd
import numpy as np
from src.api_osmr import get_matrix, parse_coordinates
from src.routing import optimize_routes, plot_route
```

# Reading data

```python
alpa = pd.read_csv("data/input/Data_Alfamart Indomaret_South Jakarta.csv")
alpa.head()
```

```python
district = alpa.groupby(['nama_kecamatan'])[['lat','lng']].median().reset_index()
```

```python
district.head(15)
```

# Rute optimal untuk mengunjungi semua kecamatan (Alfamart dan Indomart)

```python
# Extract the list of district from the "district" DataFrame
list_camat = district.nama_kecamatan

# Parse koordinat for the district
koordinat = parse_coordinates(district)

# Get the distance matrix for the district using the OSRM API
respon_jarak = get_matrix(koordinat)

# Convert the jarak to integer values (required by ortools)
jarak = np.array(respon_jarak["distances"]).astype(int)

# Set the starting nama_kecamatan for the route (change if needed)
camat_start = "Tebet"

# Find the index of the starting nama_kecamatan in the list_camat
index_depot = int(list_camat.index[list_camat == camat_start][0])

# Optimize the route using the distance matrix and starting nama_kecamatan index
rute_idx = optimize_routes(jarak, index_depot, list(list_camat))
```

```python
# Create a DataFrame with the best route
alpa_best_rute = district.iloc[rute_idx]

# Get the koordinat of the best route
koordinat = list(zip(alpa_best_rute["lat"], alpa_best_rute["lng"]))

# Plot the best route on a Folium map
map_ku = plot_route(koordinat, dist=30000, zoom_start=11)
map_ku
```

```python
# saving interactive map in html
map_ku.save('maps/map_cities_route.html')
```

# Rute optimal untuk mengunjungi 10 titik dalam 1 kelurahan

```python
# Iterate over each unique nama_kecamatan in the alpa data
for nama_kecamatan in alpa.nama_kecamatan.unique():
    print(f"Planning route for {nama_kecamatan}")

    # Select the first 10 points of sales for the current nama_kecamatan
    df_camat = alpa[alpa.nama_kecamatan == nama_kecamatan][:10].reset_index(drop=True)

    # Parse koordinat for the selected points of sales
    koordinat = parse_coordinates(df_camat)

    # Get the distance matrix for the selected points of sales using the OSRM API
    respon_jarak = get_matrix(koordinat)

    # Convert the jarak to integer values (required by ortools)
    jarak = np.array(respon_jarak["distances"]).astype(int)

    # Set the names of the points of sales for labeling the route
    nama_pos = list(df_camat.nama_kelurahan)

    # Setting the farthest point as the start point
    # Calculate the median tuple for the distance matrix
    tuples_koordinat = np.array(respon_jarak["distances"]).astype(int)
    tuple_median = tuple(np.median(np.array(tuples_koordinat), axis=0))

    # Calculate the Euclidean distance of each tuple from the median tuple
    tuple_jarak = [
        np.linalg.norm(np.array(t) - np.array(tuple_median)) for t in tuples_koordinat
    ]

    # Find the index of the tuple with the maximum distance
    index_farthest = max(enumerate(tuple_jarak), key=lambda x: x[1])[0]

    # Optimize the route using the distance matrix and farthest point index as the starting point
    rute_best = optimize_routes(jarak, index_farthest, nama_pos)[:-1]

    # Save the best route to a CSV file with the nama_kecamatan nama_kelurahan in the output folder
    df_camat.nama_kelurahan[rute_best].reset_index(drop=True).to_csv(
        f"data/output/{nama_kecamatan}_routing.csv"
    )

    # Plot the best route on the map for the nama_kecamatan "Tebet" (first nama_kecamatan in the iteration)
    if nama_kecamatan == "Mampang Prapatan":
        df_best_route = df_camat.iloc[rute_best]
        koordinat = list(zip(df_best_route["lat"], df_best_route["lng"]))
        map_ku = plot_route(koordinat, dist=2000)

# Display the final Folium map with the best route for the nama_kecamatan "Tebet"
map_ku
```
